```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
KNN_preds = np.loadtxt("prediction_results/KNN_pred_results.txt", dtype=str, skiprows=1)
KNN_combined_preds = np.loadtxt("prediction_results/KNN_combined_pred_results.txt", dtype=str, skiprows=1)

y_KNN_test = KNN_preds[:, 0]
y_KNN_pred = KNN_preds[:, 1]

y_combined_KNN_test = KNN_combined_preds[:, 0]
y_combined_KNN_preds = KNN_combined_preds[:, 1]
```

```python
accuracy = accuracy_score(y_KNN_test, y_KNN_pred)
conf_matrix = confusion_matrix(y_KNN_test, y_KNN_pred)

print(f'Accuracy: {accuracy}')
sns.heatmap(conf_matrix, 
            annot=True, # Show nums on the grd
            fmt='d', # Prevent scientific notation nums
            cmap="YlGnBu") # Color for heatmap
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
accuracy = accuracy_score(y_combined_KNN_test, y_combined_KNN_preds)
conf_matrix = confusion_matrix(y_combined_KNN_test, y_combined_KNN_preds)

print(f'Accuracy: {accuracy}')
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap="YlGnBu")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
